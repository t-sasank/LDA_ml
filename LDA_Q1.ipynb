##Importing Libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import PCA
```

##Loading Dataset(IRIS)

```python
data = load_iris()
X = data.data
y = data.target
species = data.target_names
```

##Standardisation

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

##LDA

```python
components = 2
lda = LDA(n_components=components)
X_lda = lda.fit_transform(X_std, y)
```

##Visualisation & Comparison

```python
#LDA
plt.figure(figsize=(12, 7))
plt.subplot(1, 2, 1)
for i, target in zip(range(3), species):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], label=target, marker='*')

plt.title("LDA - Iris Dataset")
plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
plt.legend()

#PCA
pca = PCA(n_components=components)
X_pca = pca.fit_transform(X_std)

plt.subplot(1, 2, 2)
for i, target_name in zip(range(3), species):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_name, marker='*')
plt.title("PCA - Iris Dataset")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()

plt.show()
```
